```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = '../results_experiments/mnist_cnum100_dist2_skew0.05_seed0/'
normal = pd.read_csv(path + 'add_0_delete_0/test_results.csv')['test_acc'][:200] 
add_2 =  pd.read_csv(path + 'add_2_delete_0/test_results.csv')['test_acc'][:200]
add_3 =  pd.read_csv(path + 'add_3_delete_0/test_results.csv')['test_acc'][:200]

add_5 =  pd.read_csv(path + 'add_5_delete_0/test_results.csv')['test_acc'][:200]
delete_3 =  pd.read_csv(path + 'add_0_delete_3/test_results.csv')['test_acc'][:200]
add_and_delete =  pd.read_csv(path + 'add_5_delete_5/test_results.csv')['test_acc'][:200]

# freq_3 =  pd.read_csv(path + 'non_iidx1_algoxfed_edgeavg_vx50_freqx3_num_edgex5_num_epochsx3_proportionx0.5.csv')['test_acc'][:200]
# freq_6 =  pd.read_csv(path + 'non_iidx1_algoxfed_edgeavg_vx50_freqx6_num_edgex5_num_epochsx3_proportionx0.5.csv')['test_acc'][:200]
# freq_9 =  pd.read_csv(path + 'non_iidx1_algoxfed_edgeavg_vx50_freqx9_num_edgex5_num_epochsx3_proportionx0.5.csv')['test_acc'][:100]

plt.xlabel("Round", fontsize = 12)
plt.ylabel("Test acc", fontsize = 12)
plt.plot(normal, label = 'Normal')
plt.plot(add_2, label = 'Add 2')
# plt.plot(add_3, label = 'Add 3')

# plt.plot(add_5, label = 'Add 5')
plt.plot(add_and_delete, label = 'Add and delete')

plt.plot(delete_3, label = 'Delete')

# plt.plot(freq_9, label ='EdgeAvg - F = 9 (81.13)'.format(max(freq_9)))

plt.legend(loc = 'lower right' ,bbox_to_anchor  = (1,-0.2))
plt.title('MNIST - test accuracy', loc = 'center', y = 1.1, fontsize = 15)
plt.show() 
```
